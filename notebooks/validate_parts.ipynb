```python
def save_lingering(search, df): 
    try:
        _sample = df[df.preceding.str.contains(f" {search} ")].sample(10).filter(['bigram_lower', 'preceding', 'adj_form_lower', 'token_str'])
    except ValueError:
        _sample = df[df.preceding.str.contains(f" {search} ")].head(10).filter(['bigram_lower', 'preceding', 'adj_form_lower', 'token_str'])
    if not _sample.empty:
        nbu.save_latex_table(_sample.style, 
                        latex_stem=f"{search}-sample", 
                        caption=f'Sample of Positive Superset hits with "{search}" in the preceding text', 
                        latex_subdir=TEX_SUB+'/lingering-potential-negs', verbose=True, text=True)
save_lingering(r"cant", ex_front)
```
