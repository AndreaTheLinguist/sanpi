````python
full_blind = adjust_am_names(
    pd.read_parquet(
        AM_DF_DIR.joinpath(
            'adv_adj/ANYdirect/extra/AdvAdj_ALL_any-direct_final-freq_min10x_extra.parq'),
        filters=[('l1', '==', LEX)], engine='pyarrow'
    ).filter(FOCUS_DICT['ALL']['adv_adj'])).assign(polarity='ANY').convert_dtypes()
full_blind.nlargest(30, 'LRC')
````

````python
neg_compose = adjust_am_names(
    pd.read_parquet
    (AM_DF_DIR.joinpath(
        'adv_adj/RBdirect/extra/AdvAdj_ALL_RBdirect_final-freq_min10x_extra.parq')
                    ).filter(FOCUS_DICT['ALL']['adv_adj'])).assign(polarity='(-)').convert_dtypes()
neg_compose
````

````python
com_compose = adjust_am_names(
    pd.read_parquet(
        AM_DF_DIR.joinpath(
            'adv_adj/not-RBdirect/extra/AdvAdj_ALL_not-RBdirect_final-freq_min10x_extra.parq')
        ).filter(FOCUS_DICT['ALL']['adv_adj'])).assign(polarity='(+)').convert_dtypes()
com_compose
````

````python
def lex_filter(compose_amdf):
    _lexed = compose_amdf.filter(regex=r'\b'+LEX+r'\b', axis=0)
    return _lexed.loc[_lexed.f>=25]

lex_neg_compose = lex_filter(neg_compose)
lex_neg_compose
````

````python
lex_com_compose = lex_filter(com_compose)
lex_com_compose
````

````python
lex_polar_compose = pd.concat([pol_compose.assign(polar_key = pol_compose.index + '=' + pol_compose.polarity.str.upper(), 
                                                  bigram=pol_compose.index.str.replace('~', '_')).set_index('bigram').reset_index().set_index('polar_key')
                               for pol_compose in (lex_com_compose, lex_neg_compose)]
                              )
lex_polar_compose 
````

````python
enhance_descrip(lex_polar_compose.select_dtypes(include='number'))
````

````python
# current_cols = lex_polar_compose.columns
def front_cols(lex_df, front:list = ['polarity', 'bigram'] + BLIND_PRIORITY):
    return lex_df.filter(front).join(lex_df.drop(columns=front))

lex_polar_compose = front_cols(lex_polar_compose)
lex_polar_compose
````
