```python
def write_set(adv_set, fstem): 
    set_list = list(adv_set)
    set_list.sort()
    Path(f'utils/{fstem}.txt').write_text('\n'.join(set_list))
```

```python
write_set(cat_def, 'adv_categorized')
write_set(ABD, 'adv_braindump')
write_set(top_adv, 'adv_most-common')
write_set(cat_def.union(ABD).union(top_adv), 'adv_anywhere')
```
