```python
def load_hit_table(adv_set): 
    
    # hit_df = pd.read_pickle(HITS_PATH).filter(regex=r'^[nab].*lower|text|str|head')
    # hit_df = pd.read_pickle(HITS_PATH)
    hit_df = pd.concat([pd.read_pickle(p).filter(regex=r'^[nab].*lower|text|str|head') 
           for i,p in enumerate(Path('/share/compling/data/sanpi/2_hit_tables/RBXadj/simple').iterdir()) if i < 4] )
    
    hit_df = hit_df.loc[hit_df.adv_form_lower.isin(adv_set), :]
    hit_df = hit_df.drop_duplicates()
    word_cols = hit_df.filter(regex=r'head|lower').columns
    hit_df[word_cols] = hit_df[word_cols].astype('category')
    return  hit_df
hits_df = load_hit_table(adv_set)
hits_df
```
