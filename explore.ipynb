This is the basic collocation counts for the negative contexts, sorted by most common in all negative contexts combined

```python
top_300_colloc = pd.read_csv('data_samples/negative_colloc-freq_top300.csv')
top_300_colloc.head()
```

10% of the sum of all negative context collocation hits is:

```python
freq10percent = top_300_colloc.iat[1,1] * 0.1
freq10percent
```

I selected only those collocations which account for 10% of the combined negative hits *on their own*. That is, their cross-context total is over the 10% mark.  

The hope is that these will skew toward negative contexts rather than positive, since they are so common in the negative contexts. This is not guaranteed though, since they could just be extremely common collocations across the board, and possibly even *more common* in positive contexts.

```python

NegTopCollocs_df = top_300_colloc[top_300_colloc.cross_context_sum >= freq10percent]
NegTopCollocs_df = NegTopCollocs_df.rename(columns={'cross_context_sum':'NEG_sum'}).loc[1:,:]
NegTopCollocs = NegTopCollocs_df.colloc
NegTopCollocs_df
```

I can then select only these collocation from the positive frequency table as well. 

First, here are the most frequent collocations in positive contexts (already see some possibility of level distribution...)

```python
pos_collocs = pd.read_pickle('positive-contexts_colloc-freq-table.pkl.gz')
pos_collocs = pos_collocs.rename(columns={'be-ADV-ADJ_positive':'POS_sum'})
pos_collocs
```

```python
POS_top_300_colloc = pd.read_csv('data_samples/positive_colloc-freq_top300.csv')
POS_top_300_colloc.head(11)
```

Comparing the collocation frequencies between context type, it seems there may indeed be some collocations that skew strongly negative:

```python
pos_for_compare = pos_collocs.loc[NegTopCollocs,:]


neg_for_compare = NegTopCollocs_df.set_index('colloc')

compare = neg_for_compare.join(pos_for_compare)
compare = compare[['NEG_sum','POS_sum']].fillna(0).convert_dtypes()

# compare = neg_for_compare.assign(positive_total=pos_for_compare)
compare
```

## Divergence of positive distribution from negative:

```python
round(entropy(compare.POS_sum.to_list(), compare.NEG_sum.to_list()),3)
```

## Divergence of negative distribution from positive

**This is the highest divergence value for this subset of data**

```python
round(entropy(compare.NEG_sum.to_list(), compare.POS_sum.to_list()),3)
```

I think the following is indicative of why the divergence values weren't informative. The disparity in the amount of data is 

```python
unk_collocs = pd.read_pickle('test-contexts_colloc-freq-table.pkl.gz')
unk_collocs = unk_collocs.assign(UNK_sum=unk_collocs.sum(axis=1))
unk_collocs.sort_values('UNK_sum',ascending=False)
```

```python

unk_collocs_overlap = unk_collocs[unk_collocs.index.isin(NegTopCollocs)]
unk_collocs_overlap
```

```python
compare_unk = compare.assign(UNK_sum = unk_collocs_overlap.UNK_sum).fillna(0).convert_dtypes()
compare_unk
```

Now we have to do some smoothing bc 0s will break the algorithm. 

```python
compare_unk = compare_unk.add(1)
compare_unk
```

```python
posdist = np.array(compare_unk.POS_sum.to_list())
negdist = np.array(compare_unk.NEG_sum.to_list())
unkdist = np.array(compare_unk.UNK_sum.to_list())
```

## Divergence of unknowns from negative distribution

```python
round(entropy(unkdist, negdist),3)
```

## Divergence of unknowns from positive distribution

```python
round(entropy(unkdist, negdist),3)
```
