```python
import pandas as pd

context=pd.read_csv('/share/compling/data/sanpi/corpora_shortcuts/puddin/Pcc11.conll/subset_bigram/BIGRAM.pcc_eng_11-088.context.psv',sep="|")
```

```python
def _make_categoricals(df: pd.DataFrame,
                       cat_cols= None,
                       cat_suff: tuple = ("code", "name", "path", "stem")):
    cat_cols = cat_cols if cat_cols else df.columns.str.endswith(cat_suff)
    df.loc[:, cat_cols] = df.loc[:, cat_cols].astype("category")
    return df
```

```python
context.info()
context = context.astype('string') #convert_dtypes()
context['sent_int'] = pd.to_numeric(context.sent_int, downcast='unsigned')
context = _make_categoricals(context, ['conllu_id', 'doc_id'])
context = context.set_index('sent_id')
context.info()
```

```python
nac = context.loc[context.prev_id.isna(), :]
nac = nac.loc[nac.sent_int != 1, :]
nac
```
