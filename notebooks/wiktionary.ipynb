```python
import requests
from pprint import pprint

def request_url(bigram): 
    return f'https://en.wiktionary.org/api/rest_v1/page/definition/{bigram}?redirect=false'

def lexicalized(response): 
    return 'en' in response.keys()

# from os import system
# system(f"curl -X 'GET' {request_url('scantily_clad')} -H 'accept: application/json; charset=utf-8; profile=\"https://www.mediawiki.org/wiki/Specs/definition/0.8.0\"'")

BIGRAMS = ('very_good', 'pretty_tall', 'exactly_sure', 'too_shabby', 'scantily_clad', 'brand_new', 'terminally_ill', 'immediately_available')
```

```python
entries = {}
for bigram in BIGRAMS:
    response = requests.get(request_url(bigram))
    entries[bigram] = response.json()
    print(f'## {bigram}')
    pprint(response.json())
```

```python
import pandas as pd
df = pd.Series(entries).to_frame('response')
df.index.name = 'bigram'
df['lexical'] = df.response.apply(lexicalized)
df
```

```python
print(df.lexical.sort_values().astype('boolean').to_markdown())
```

| bigram                | lexical   |
|:----------------------|:----------|
| pretty_tall           | False     |
| exactly_sure          | False     |
| too_shabby            | False     |
| immediately_available | False     |
| very_good             | True      |
| scantily_clad         | True      |
| brand_new             | True      |
| terminally_ill        | True      |


```python
indenter = "_\n  + _"
for g, words in df.groupby('lexical'): 
    kind = "Lexicalized"
    if not g: 
        kind = f'Not {kind}'
    print(f'\n### {kind}\n\n  + _{indenter.join([w.replace("_", " ") for w in words.index.to_list()])}_')
```
