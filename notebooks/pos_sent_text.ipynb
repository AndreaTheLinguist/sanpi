```python
from pathlib import Path
import pandas as pd
NEG_LEMMAS = ("hardly", "scarcely", "never", "rarely", "barely", "seldom", "no", "nothing", "none", "nobody", "neither", "without", "few", "nor")
Path.cwd()
```

```python
pos_csv = Path('exactly_out/pos_sentences.csv').resolve()
df = pd.read_csv(pos_csv).convert_dtypes().set_index('sent_id')
# df.sort_values('adj_neg_ratio', ascending=False)
df.sort_values('adj_neg_ratio', ascending=False)[['adj_lemma', 'text_window', 'conllu_id']]
```

```python
df = df.assign(doc_id = df.index.str.rsplit('_', 1).str.get(0), # type: ignore
               sent_ix = df.index.str.rsplit('_', 1).str.get(1), 
               neg_likelihood = df.adj_neg_ratio.round(1)) # type: ignore

for neglem in NEG_LEMMAS:
    # neg_sents_path = pos_csv.with_name(f'{pos_csv.stem}_{neglem}.csv')
    # print(f'{neglem} "pos" sentences saved to {neg_sents_path}')
    df.loc[:, neglem] = df.sent_text.str.contains(f' {neglem} ')
df
```

```python
df.loc[:, 'nt'] = df.sent_text.str.contains("n't")
df.loc[:, 'hardly':]
```
