# Exploring hits from new pattern approach, as they currently stand

```python
import pandas as pd
from pprint import pprint
from pathlib import Path
Patdir = Path('./Pat')
hitsdir = Path('./hits')
```

## Current pattern specifications:

```python
for pf in Patdir.glob('**/*.pat'):
    print(pf.relative_to('.'), ':\n------------')
    print(pf.read_text(), '\n')
```

## Loading all hits from sample conllu files for above .pat files

```python
hits = pd.DataFrame()
for hf in hitsdir.glob('**/*hits.csv'):

    csv_df = pd.read_csv(hf)

    hits = pd.concat([hits, csv_df], ignore_index=True)
```

```python
pprint(hits.columns.sort_values().to_list())
```
